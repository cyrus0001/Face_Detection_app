## Testing Camera 

```python
#Importing Numpy library for numerical operations
import numpy as np 

# Importing OpenCV library for computer vision tasks
import cv2 
```

```python
# Initializing video capture from the default camera (index 0)
cap = cv2.VideoCapture(0)  
cap.set(3, 640) 
cap.set(4, 480) 

while True: 
    ret, frame = cap.read()  # Capturing a frame from the video capture
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Converting the frame to grayscale
    
    # Displaying the original frame and the grayscale frame in separate windows
    cv2.imshow('frame',frame)
    cv2.imshow('gray', gray)   
    
    k = cv2.waitKey(30) & 0xff  # Waiting for a key press for 30 milliseconds and storing the key value
    if k == 27:  # Checking if the pressed key is 'ESC' (ASCII code 27)
        break 

cap.release()  # Releasing the video capture object, releasing the camera resource
cv2.destroyAllWindows()  # Closing all OpenCV windows
```
